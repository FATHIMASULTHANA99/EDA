### can we combine these 2 dataframes? how should we combine?
An. Merge function

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
# lets explore our final_df dataset

final_df.columns
```

```python
country_counts=final_df.Country.value_counts()
country_counts
```

```python
# plotting country_counts
plt.figure(figsize=(12, 4))
country_counts.plot(kind='bar')
plt.title('Country Distribution')
plt.xlabel('Country')
plt.ylabel('Count')
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_values=final_df.Country.value_counts().values
country_values
```

```python
#plotting
plt.pie(country_values,labels=country_names)
plt.show()
```

```python
# this plot is not good enough so we are taking only top3 countries to plot
plt.pie(country_values[:3],labels=country_names[:3],autopct='%.2f %%')
```

#### observation:
zomatos maximum record or transactions from India
after that USA and UK

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
rating=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
rating
```

#### obseravtions
1.when rating in between 4.5 to 4.9 ---> Excellent

2. when rating in between  4.0 to 4.4---> Very Good

3.when rating in between 3.5 to 3.9---> Good

 etc etc...

```python
# plotting relation between aggregate_rating and rating_count
plt.figure(figsize=(12,5))
sns.barplot(x='Aggregate rating',y='Rating count',data=rating)
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Aggregate rating',y='Rating count',data=rating,hue='Rating color')
```

```python
plt.figure(figsize=(12,5))
# the color is not matching , so 
sns.barplot(x='Aggregate rating',y='Rating count',data=rating,hue='Rating color',palette=['violet','red','orange','yellow','green','green'])
```

#### observations
1.not rated count is very high

2. max number of ratings are between 2.6 to 3.9

```python
## count plot
sns.countplot(x='Rating color',data=rating,palette=['violet','red','orange','yellow','green','green'])
```
